#**Import The Necessary Libraries**

```python
import pandas as pd
import pytrends
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trends = TrendReq()
```

#**Now Let's Find Countries Where Certain Keywords Are Searched The Most**

```python
#add your keyword here
trends.build_payload(kw_list=['machine learning']) 
data = trends.interest_by_region()

data = data.sort_values(by="machine learning", ascending=False) 
data = data.head(20)

print(data)
```

```python
data.reset_index().plot(x="geoName", y="machine learning", figsize=(15,10), kind="bar", color = 'green')
plt.style.use('fivethirtyeight')
plt.show()
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['machine learning'])

data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(15, 10))

data['machine learning'].plot()

plt.title("Total Google Searches for keyword: 'Machine Learning'", fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')

plt.style.use('fivethirtyeight')
plt.show()
```

#**Historical Interest Overtime**

```python
keywrd_list = ["pakistan"]
data = trends.get_historical_interest(keywrd_list, year_start=2021, month_start=1, 
                                      day_start=1, hour_start=0, year_end=2021, month_end=3, 
                                      day_end=1, hour_end=0, cat=0, geo='', gprop='', sleep=0)

data = data.sort_values(by="pakistan", ascending=False) 
data = data.head(30)

print(data)
```

```python
data.reset_index().plot(x="date", y="pakistan", 
                        figsize=(15,6), kind="bar", color='green')
plt.style.use('fivethirtyeight')
plt.show()
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['python programming'])

data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(15, 6))

data['python programming'].plot()
plt.title("Total Google Searches for keyword: 'Python Programming' overtime", fontweight='bold')

plt.xlabel('Year')
plt.ylabel('Total Count')

plt.show()
```
