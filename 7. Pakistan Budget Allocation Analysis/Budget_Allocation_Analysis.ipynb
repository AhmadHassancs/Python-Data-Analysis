```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("budget-allocation.csv")
data
```

```python
data['Allocated Budget (Rs. In Millions)'] = pd.to_numeric(data['Allocated Budget (Rs. In Millions)'])
```

```python
data['Province'] = data['Province'].astype('category')
```

```python
data['Sector/Sub-Sector'] = data['Sector/Sub-Sector'].astype('category')
data['Sector/Sub-Sector']
```

```python
data['Province']
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
data=data.dropna()
```

```python
data.isna().sum()
```

# Population in each Province

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Province', y='Population', data=data)
```

# Budget Allocation to different sectors

```python
plt.figure(figsize=(15,8))
sns.barplot(x='Sector/Sub-Sector', y='Allocated Budget (Rs. In Millions)', data=data)
```

# Total Budget Allocated to each province so far

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Province', y='Allocated Budget (Rs. In Millions)', data=data, palette=['green','blue','orange','yellow'])
sns.set_palette("Paired")
```

# **Distribution of The Budget For Top 10 Ministries**

```python
data = data.sort_values(by='Allocated Budget (Rs. In Millions)',ascending=False)
data.head(10)
```

```python
new_data=data.head(10)
df = new_data["Allocated Budget (Rs. In Millions)"]
labels = new_data["Sector/Sub-Sector"]
plt.figure(figsize=(10,9))
plt.pie(df, labels=labels, autopct='%1.1f%%', startangle=45, pctdistance=0.7, shadow =True)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Distribution of The Budget For Top 10 Sectors", fontweight='bold')
plt.show()
```

# Distribution of The Budget For Last 10 Ministries

```python
data = data.sort_values(by='Allocated Budget (Rs. In Millions)',ascending=True)
new_data = data.head(10)
new_data
```

```python
df = new_data["Allocated Budget (Rs. In Millions)"]
labels = new_data["Sector/Sub-Sector"]
plt.figure(figsize=(10,9))
plt.pie(df, labels=labels, autopct='%1.1f%%', startangle=0, pctdistance=0.7, shadow =True)
central_circle = plt.Circle((0, 0), 0.8, color='white') #change the 0.8 to see changes to the circle
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Distribution of The Budget For Last 10 Ministries", fontweight='bold')
plt.show()
```
