## Imports

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Reading the Data

```python
df = pd.read_csv('all.csv')
```

```python
df.head()
```

### Data Exploration

```python
# Statistics of the data
df.describe()
```

```python
#Boxplot for Months since Last Donation
plt.figure(figsize=(8,8)) 
sns.boxplot(y="Months since Last Donation",data=df)
```

We see from the above boxplot that the maximum people have donated blood in nearby 10 months.

```python
plt.figure(figsize=(8,8)) 
sns.boxplot(y="Months since First Donation",data=df)
```

```python
sns.pairplot(data=df)
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(data=df.corr(), cmap='coolwarm', linewidths = 0.1,fmt = ".2f",annot=True)
```

```python
plt.figure(figsize=(10,7))
sns.barplot(x='Months since Last Donation', y='Months since First Donation', data=df)
```

```python
plt.figure(figsize=(10,7))
sns.barplot(x='Months since Last Donation', y='Months since First Donation',hue='Number of Donations', data=df)
```
